# PGC - Demo Notebook: Data Visualization and Analysis Tools

```python
import os
import sys

sys.path.append(os.path.abspath(".."))  # add path
from sklearn.datasets import fetch_california_housing, load_diabetes, load_iris

from utils.interactive_dataframe import InteractiveDataFrame as iDataFrame
```

# Diabetes Dataset 

```python
diabetes = load_diabetes(return_X_y=True, as_frame=True)[0]
diabetes = iDataFrame(diabetes)
diabetes.render_with_widgets()
```

Plot specific visualizations of interative panel

```python
save_dict = {"hist": [0], "box": [1, 2, 3, 4], "scatter":[2,4]}
selector_plt = diabetes.get_viz(save_dict)
iDataFrame.plt_from_selector(selector_plt)
```

Or Save in a file so you can load it later 

```python
# Saving viz in external file
diabetes.save_plt(save_dict, fname="diabetes_plot_selection.pickle")

# Loading from file or object
selector_plt = iDataFrame.load_plt("diabetes_plot_selection.pickle")

iDataFrame.plt_from_selector(selector_plt)
```

### Plotting Ad Hoc
Through the below methods it's possible, for a specific visualization type and/or specific columns, to plot the top 5 or all columns. 

```python
diabetes.plt("hist", diabetes.columns[0:4])
```

```python
diabetes.plt5("box")
```

```python
diabetes.plt_all("hist")
```

```python
diabetes.plt_topn('scatter', 10)
```

# Iris Dataset

```python
iris = load_iris(return_X_y=True, as_frame=True)[0]
iris = iDataFrame(iris)
iris.render_with_widgets()
```

```python
save_dict = {"scatter": [0, 1, 2], "box": [2]}
selector_plt = iris.get_viz(save_dict)
iDataFrame.plt_from_selector(selector_plt)
```

```python
iris.plt("box", ['sepal length (cm)', 'sepal width (cm)'])
iris.plt5("box")
iris.plt_all("scatter")
iris.plt_topn("hist", 10)
```

# Housing Dataset 

```python
housing = fetch_california_housing(return_X_y=True, as_frame=True)[0]
housing = iDataFrame(housing)

housing.render_with_widgets()
```

```python
## Saving selected plots
save_dict = {"scatter": [0, 1, 2, 3, 4], "box": [0, 1, 2, 3, 4], "hist": [0, 1, 2, 3, 4]}
selector_plt = housing.get_viz(save_dict)
iDataFrame.plt_from_selector(selector_plt, per_row=5)
```

```python
housing.plt("scatter", housing.columns[0:3])
housing.plt5("box")
housing.plt_all("hist")
```

```python
from utils.interactive_dataframe import InteractiveDataFrame as iDataFrame
```

```python
from sklearn.datasets import load_iris, fetch_california_housing, load_diabetes

iris = load_iris(return_X_y=True, as_frame=True)[0]
housing = fetch_california_housing(return_X_y=True, as_frame=True)[0]
diabetes = load_diabetes(return_X_y=True, as_frame=True)[0]

iris = iDataFrame(iris)
housing = iDataFrame(housing)
diabetes = iDataFrame(diabetes)
```

```python
iris.render_with_widgets()
```

```python
## Saving selected plots
save_dict = {'scatter': [0, 1, 2], 'box': [2]}
selector_plt = iris.get_viz(save_dict)
iDataFrame.plt_from_selector(selector_plt)
```

Or Save in a file so you can load it later 

```python
iris.save_plts(save_dict, fname="iris_plot_selection.pickle")
selector_plt = iDataFrame.load_plts("iris_plot_selection.pickle")

iDataFrame.plt_from_selector(selector_plt)
```

```python
iris.plt_all("scatter")
```

```python
housing.render_with_widgets()
```

```python
## Saving selected view
selector_plt = housing.get_viz({'scatter': [0], 'box': [1,2, 3, 4]})
print(selector_plt)
iDataFrame.plt_from_selector(selector_plt)
```

```python
diabetes.render_with_widgets()
```
